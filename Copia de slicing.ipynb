### Slicing de Filas

Puedes seleccionar rangos de filas utilizando `iloc` (por índice numérico) o `loc` (por etiqueta de índice, si la tuvieras).

**Ejemplo con `iloc` (primeras 5 filas):**

```python
print("Primeras 5 filas del DataFrame (iloc):")
display(df.iloc[0:6])
```

**Ejemplo con `iloc` (filas 10 a 14):**

```python
print("Filas 10 a 14 del DataFrame (iloc):")
display(df.iloc[10:15])
```

### Slicing de Columnas

Puedes seleccionar una o varias columnas utilizando sus nombres.

**Ejemplo seleccionando la columna 'App' y 'Category':**

```python
print("Columnas 'App' y 'Category':")
display(df[['App', 'Category']].head())
```

### Slicing Combinado (Filas y Columnas)

También puedes combinar el slicing para seleccionar un rango de filas y un rango de columnas.

**Ejemplo: primeras 3 filas y columnas 'App' y 'Rating':**

```python
print("Primeras 3 filas y columnas 'App', 'Rating':")
display(df.loc[0:2, ['App', 'Rating']])
```

### Slicing con Índices Negativos (Últimas Filas)

Puedes usar índices negativos con `iloc` para seleccionar filas desde el final del DataFrame. Por ejemplo, `df.iloc[-5:]` te dará las últimas 5 filas.

```python
print("Últimas 5 filas del DataFrame (iloc con índices negativos):")
display(df.iloc[-5:])
```

### Slicing de Filas y Columnas No Consecutivas (con Listas de Índices)

Si necesitas seleccionar filas y/o columnas que no son contiguas, puedes pasar listas de índices a `iloc`. Por ejemplo, para seleccionar las filas 0, 2 y 5, y las columnas 1 y 3:

```python
print("Filas 0, 2, 5 y columnas 1, 3 (iloc con listas de índices):")
display(df.iloc[[0, 2, 5], [1, 3]])
```

### Slicing Condicional (Filtrado por Valor de Columna)

Una de las formas más potentes de seleccionar datos es usando condiciones lógicas. Con `loc`, puedes filtrar filas basándote en el valor de una columna y luego seleccionar las columnas que te interesan. Por ejemplo, para obtener las apps con una 'Rating' mayor a 4.5, mostrando solo las columnas 'App' y 'Rating':

```python
print("Apps con Rating > 4.5 (slicing condicional):")
display(df.loc[df['Rating'] > 4.5, ['App', 'Rating']].head())
```

### Slicing `[:]` (Todas las Filas y Todas las Columnas)

El formato `df.iloc[:, :]` selecciona todas las filas y todas las columnas de tu DataFrame. Es equivalente a mostrar el DataFrame completo, aunque usaremos `.head()` para ver solo las primeras filas por eficiencia.

```python
print("Todas las filas y columnas (primeras 5 filas):")
display(df.iloc[:, :].head())
```

### Slicing `[3, :]` (Fila en Índice 3 y Todas las Columnas)

El formato `df.iloc[3, :]` selecciona la fila específica en el índice numérico `3` y todas sus columnas correspondientes. Recuerda que el indexado es base 0, por lo que `3` es la cuarta fila.

```python
print("Fila en el índice 3 y todas sus columnas:")
display(df.iloc[3, :])
```

### Slicing `[:, 9]` (Todas las Filas y Columna en Índice 9)

El formato `df.iloc[:, 9]` selecciona todas las filas y la columna específica que se encuentra en el índice numérico `9`. Veremos las primeras 5 filas de esta columna.

```python
print("Todas las filas de la columna en el índice 9 (primeras 5 filas):")
display(df.iloc[:, 9].head())
```
